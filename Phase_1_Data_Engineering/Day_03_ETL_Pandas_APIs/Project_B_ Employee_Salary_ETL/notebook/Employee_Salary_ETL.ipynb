**Read Excel**

```python
df = pd.read_excel(
    "employee_salary.xlsx"
)
```

**Remove duplicates**

```python
df.drop_duplicates(
    inplace=True
)
```

**Missing Value Imputation**

```python
df["Basic_Salary"] = df["Basic_Salary"].fillna(
    df["Basic_Salary"].mean()
)
```

**Derived fields**

**Gross Salary**

```python
df["Gross_Salary"] = (
    df["Basic_Salary"]
    +
    df["Basic_Salary"]*0.20
    +
    df["Basic_Salary"]*0.10
)
```

**Tax**

```python
df["Tax"] = (
    df["Gross_Salary"]*0.10
)
```

**Net salary**

```python
df["Net_Salary"] = (
    df["Gross_Salary"]
    -
    df["Tax"]
)
```

**Summary Report**

```python
summary = pd.DataFrame({
    "Metric":[
        "Employees",
        "Average Salary",
        "Highest Salary",
        "Lowest Salary"
    ],
    "Value":[
        len(df),
        df["Net_Salary"].mean(),
        df["Net_Salary"].max(),
        df["Net_Salary"].min()
    ]
})
```

**Export**

```python
df.to_excel(
    "cleaned_employee_salary.xlsx",
    index=False
)
```

```python
summary.to_csv(
    "summary_report.csv",
    index=False
)
```
