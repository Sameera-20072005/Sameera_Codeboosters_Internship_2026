# **PROJECT A — Weather Data ETL**

**Install libraries**

```python
import requests
import pandas as pd
```

**Extract data**

```python
url = "https://api.open-meteo.com/v1/forecast"

params = {
    "latitude": 13.15,
    "longitude": 80.25,
    "current_weather": True
}

response = requests.get(url, params=params)

weather_json = response.json()
print(weather_json)
```

**Save Raw JSON**

```python
import json

with open("raw_weather.json","w") as f:
    json.dump(weather_json,f)
```

**Normalise JSON**

```python
df = pd.json_normalize(
    weather_json["current_weather"]
)
display(df)
```

**Cleaning**

```python
df.fillna(0,inplace=True)
```

```python
df["temperature"] = df["temperature"].astype(float)
```

```python
df.drop_duplicates(inplace=True)
```

**Export**

```python
df.to_csv(
    "cleaned_weather.csv",
    index=False
)
```
